```python
# ...existing code...
import matplotlib.pyplot as plt
import numpy as np

# Data 3 run untuk Gaussian Naive Bayes (GNB) â€” dalam detik
methods = ['Grid Search CV', 'Random Search CV', 'Halving Random Search CV']

waktu_run_1 = [2.22, 0.87, 204.32]
waktu_run_2 = [2.91, 0.35, 200.98]
waktu_run_3 = [2.13, 0.34, 206.03]

# Hitung rata-rata dan standar deviasi
mean_times = np.mean([waktu_run_1, waktu_run_2, waktu_run_3], axis=0)
std_times = np.std([waktu_run_1, waktu_run_2, waktu_run_3], axis=0)

# Gunakan detik (GNB cepat)
mean_seconds = mean_times
std_seconds = std_times

# Plot horizontal bar
y = np.arange(len(methods))
fig, ax = plt.subplots(figsize=(8, 5))

bars = ax.barh(y, mean_seconds, xerr=std_seconds, capsize=8,
               color=['#1f77b4', '#ff7f0e', '#2ca02c'],
               edgecolor='black', linewidth=1.0, alpha=0.9)

# Tulis nilai di ujung setiap batang horizontal
for bar, mean, std in zip(bars, mean_seconds, std_seconds):
    width = bar.get_width()
    ax.text(width + max(std_seconds)*0.05, bar.get_y() + bar.get_height() / 2,
            f'{mean:.2f} Â± {std:.2f} detik',
            va='center', ha='left', fontweight='bold', fontsize=10)

ax.set_yticks(y)
ax.set_yticklabels(methods)
ax.set_xlabel('Waktu Komputasi Rata-rata (detik)',
              fontsize=12, fontweight='bold')
ax.set_title('Perbandingan Waktu Komputasi Optimasi Hyperparameter\nModel Gaussian Naive Bayes (GNB)\n(Rata-rata dari 3 kali eksekusi)',
             fontsize=13, fontweight='bold', pad=10)
ax.set_xlim(0, max(mean_seconds + std_seconds) * 1.15)
ax.grid(True, axis='x', linestyle='--', alpha=0.6)
ax.invert_yaxis()

plt.tight_layout()
plt.savefig('waktu_komputasi_gnb.png', dpi=300, bbox_inches='tight')
plt.show()
# ...existing code...
```
