```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style and figure size
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Create the data
iterations = np.arange(10)
num_candidate_models = [10, 8, 4, 2, 1, 0.5, 0.25, 0.125, 0.0625, 0.01]
num_training_cases = num_candidate_models.copy()
num_training_cases.reverse()

# Create a DataFrame
data = {
    'Iteration': iterations,
    'Number of Candidate Models': num_candidate_models,
    'Number of Training Cases': num_training_cases
}
df = pd.DataFrame(data)

# Create the plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Iteration', y='Number of Candidate Models',
             label='Kombinasi Hyperparameter', color='blue')
sns.lineplot(data=df, x='Iteration', y='Number of Training Cases',
             label='Train Set', color='red')

# Customize the plot
plt.title('Ilustrasi Succesive halving', fontsize=14)
plt.xlabel('Iterasi', fontsize=12)
plt.ylabel('', fontsize=12)
plt.legend()

plt.tight_layout()
plt.show()
```
